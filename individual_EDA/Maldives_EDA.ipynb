```python
def iqr_flags(s, k=1.5):
    q1, q3 = s.quantile([0.25,0.75])
    iqr = q3-q1
    return ((s < q1 - k*iqr) | (s > q3 + k*iqr))

outlier_report = {}
for c in num_cols:
    flags = iqr_flags(df[c].dropna())
    outlier_report[c] = int(flags.sum())
outlier_report
```
