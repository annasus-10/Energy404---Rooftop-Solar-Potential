```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_samarkand = pd.read_csv("dataset/samarkand.csv")
```

```python
df_samarkand.head()
```

```python
df_samarkand.describe()
```

```python
df_samarkand.info()
```

```python
df_samarkand=df_samarkand.drop('Unit_installation_price', axis="columns")
```

```python
df_samarkand=df_samarkand.drop("Comment", axis='columns')
```

```python
df_samarkand=df_samarkand.drop("uuid", axis='columns')
```

```python
df_samarkand.count()
```

```python
df_samarkand.head()
```

```python
df_samarkand.rename(columns = {'Surface_area':'surface_area', 
                     'Potential_installable_area':'installable_area', 
                     'Peak_installable_capacity':'capacity', 
                     'Energy_potential_per_year':'energy_year',
                     'Assumed_building_type':'building', 
                     'Estimated_tilt':'tilt', 
                     'Estimated_building_height':'height', 
                     'Estimated_capacity_factor':'capacity_factor'}, inplace = True)
```

```python
df_samarkand.columns
```

```python
df_samarkand['height'].isna().sum()
```

```python
df_samarkand[df_samarkand['height'].isna()]
```

```python
mean_height = df_samarkand[df_samarkand['building'] == 'single family residential']['height'].mean()
median_height = df_samarkand[df_samarkand['building'] == 'single family residential']['height'].median()

print("Mean height:", mean_height)
print("Median height:", median_height)
```

```python
df_samarkand['height']=df_samarkand['height'].fillna(mean_height)
```

```python
df_samarkand['height'].isna().sum()
```

```python
df_samarkand['capacity_factor'].isna().sum()
```

```python
df_samarkand.count()
```

```python
df_samarkand[df_samarkand['capacity_factor'].isna()]
```

```python
df_samarkand = df_samarkand.dropna(subset=['capacity_factor'])
```

```python
df_samarkand['capacity_factor'].isna().sum()
```

```python
df_samarkand.count()
```

```python
df_samarkand.head()
```

```python
sns.countplot(df_samarkand, x = 'building')
```

```python
sns.boxplot(x = df_samarkand["building"], y = df_samarkand["capacity"])
plt.ylabel("Peak Capacity")
plt.xlabel("Building Type")
```

```python
sns.scatterplot(x = df_samarkand['installable_area'], y = df_samarkand['capacity'], hue=df_samarkand['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Installable Area of Rooftop")
```

```python
sns.scatterplot(x = df_samarkand['tilt'], y = df_samarkand['capacity'], hue=df_samarkand['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Tilt Angle")
```

```python
sns.scatterplot(x = df_samarkand['building'], y = df_samarkand['installable_area'], hue=df_samarkand['tilt'])
plt.ylabel("Installable Area")
plt.xlabel("Building Type")
```

```python
sns.scatterplot(x = df_samarkand['height'], y = df_samarkand['capacity'], hue=df_samarkand['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Building Height")
```
