```python
df_vincent=df_vincent.drop("Comment", axis='columns')
```

```python
df_vincent.rename(columns = {'Surface_area':'surface_area', 
                     'Potential_installable_area':'installable_area', 
                     'Peak_installable_capacity':'capacity', 
                     'Energy_potential_per_year':'energy_year',
                     'Assumed_building_type':'building', 
                     'Estimated_tilt':'tilt', 
                     'Estimated_building_height':'height', 
                     'Estimated_capacity_factor':'capacity_factor'}, inplace = True)
```

```python
mean_height = df_vincent['height'].mean()
median_height =df_vincent['height'].median()

print("Mean height:", mean_height)
print("Median height:", median_height)
```

```python
df_vincent['height']=df_vincent['height'].fillna(median_height) 
```

```python
sns.countplot(df_vincent, x = 'building')
```

```python
sns.boxplot(x = df_vincent["building"], y = df_vincent["capacity"])
plt.ylabel("Peak Capacity")
plt.xlabel("Building Type")
```

```python
sns.scatterplot(x = df_vincent['installable_area'], y = df_vincent['capacity'], hue=df_vincent['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Installable Area of Rooftop")
```

```python
sns.scatterplot(x = df_vincent['tilt'], y = df_vincent['capacity'], hue=df_vincent['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Tilt Angle")
```

```python
sns.scatterplot(x = df_vincent['building'], y = df_vincent['installable_area'], hue=df_vincent['tilt'])
plt.ylabel("Installable Area")
plt.xlabel("Building Type")
```

```python
sns.scatterplot(x = df_vincent['height'], y = df_vincent['capacity'], hue=df_vincent['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Building Height")
```
