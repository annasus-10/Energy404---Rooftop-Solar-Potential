```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_almaty = pd.read_csv("dataset/almaty.csv")
```

```python
df_almaty.head()
```

```python
df_almaty.info()
```

```python
df_almaty=df_almaty.drop('Unit_installation_price', axis="columns")
```

```python
df_almaty=df_almaty.drop("Comment", axis='columns')
```

```python
df_almaty=df_almaty.drop("uuid", axis='columns')
```

```python
df_almaty.count()
```

```python
df_almaty.head()
```

```python
df_almaty.rename(columns = {'Surface_area':'surface_area', 
                     'Potential_installable_area':'installable_area', 
                     'Peak_installable_capacity':'capacity', 
                     'Energy_potential_per_year':'energy_year',
                     'Assumed_building_type':'building', 
                     'Estimated_tilt':'tilt', 
                     'Estimated_building_height':'height', 
                     'Estimated_capacity_factor':'capacity_factor'}, inplace = True)
```

```python
df_almaty.columns
```

```python
df_almaty['height'].isna().sum()
```

```python
df_almaty[df_almaty['height'].isna()]
```

```python
mean_height = df_almaty['height'].mean()
median_height =df_almaty['height'].median()

print("Mean height:", mean_height)
print("Median height:", median_height)
```

```python
df_almaty['height'].describe()
```

```python
df_almaty[df_almaty['height'] < 0]
```

```python
df_almaty.loc[df_almaty['height'] < 0, 'height'] = np.nan
```

```python
df_almaty.loc[df_almaty['height'] == 0, 'height']
```

```python
df_almaty.loc[df_almaty['height'] == 0, 'height'] = np.nan
```

```python
df_almaty['height'].describe()
```

```python
df_almaty['height']=df_almaty['height'].fillna(median_height)
```

```python
df_almaty['height'].isna().sum()
```

```python
df_almaty.count()
```

```python
df_almaty['capacity_factor'].describe()
```

```python
cf_mean=df_almaty['capacity_factor'].mean()
cf_median=df_almaty['capacity_factor'].median()
print("Mean :", cf_mean)
print("Median :", cf_median)
```

```python
df_almaty['capacity_factor']=df_almaty['capacity_factor'].fillna(cf_mean) #mean and median are not that different 
```

```python
sns.scatterplot(x = df_almaty['installable_area'], y = df_almaty['capacity'], hue=df_almaty['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Installable Area of Rooftop")
```

```python
sns.scatterplot(x = df_almaty['tilt'], y = df_almaty['capacity'], hue=df_almaty['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Tilt Angle")
```

```python
sns.scatterplot(x = df_almaty['building'], y = df_almaty['installable_area'], hue=df_almaty['tilt'])
plt.ylabel("Installable Area")
plt.xlabel("Building Type")
```

```python
sns.scatterplot(x = df_almaty['height'], y = df_almaty['capacity'], hue=df_almaty['building'])
plt.ylabel("Peak Capacity")
plt.xlabel("Building Height")
```
